```python
##creating different functions for each operation
def add_numbers(num1, num2): ##addition, not order dependent
    return num1 + num2

def subtract_numbers(num1, num2): ##subtraction, second number is subtracted from first
    return num1 - num2

def multiply_numbers(num1, num2): #multiplication, not order dependent
    return num1 * num2

def divide_numbers(num1, num2): #division, second number is divided from first
    return num1 / num2

def exponents_numbers(num1, num2): ##exponents, first number is the base, second is exponent
    return num1 ** num2

def square_roots(num1, num2): ##square roots, first number is the base, second is the degree of rooting
    return num1 ** (1/num2) ##to get the square root, the second number should be 2.


while True:
    print("Welcome to the calculator program!")
    print("Select operation:") ##numbers 1-6 are assigned to operations
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exponents")
    print("6. Square Roots")

    choice = input("Enter choice(1/2/3/4/5/6): ") ##user decides the operation

    if choice in ('1', '2', '3', '4', '5', '6'):
        num1 = float(input("Enter first number: ")) ##takes the user's first number input and converts to a decimal
        num2 = float(input("Enter second number: ")) ##takes the user's second number input and converts to a decimal

        if choice == '1': ##adds
            print(num1, "+", num2, "=", add_numbers(num1, num2))

        elif choice == '2': ##subtracts
            print(num1, "-", num2, "=", subtract_numbers(num1, num2))

        elif choice == '3': ##multiplies
            print(num1, "*", num2, "=", multiply_numbers(num1, num2))

        elif choice == '4': ##divides
            print(num1, "/", num2, "=", divide_numbers(num1, num2))

        elif choice == '5': ##exponents
            print(num1, "**", num2, "=", exponents_numbers(num1, num2))

        elif choice == '6': ##square roots
            print(num1, "**(1/", num2, ")=", square_roots(num1, num2))

    continue_calc = input("Do you want to continue? (yes/no): ") ##asks if user wants to continue
    if continue_calc.lower() != 'yes':
        print("Thank you for using the calculator program!")
        break




#this code could be improved to process logarithmic functions,, or even process 3 or more numbers
```
